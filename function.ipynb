# 文字列のパターンマッチをする関数
文字列中の特定のパターンを見つけ出す関数


``` python

import re # パターンマッチライブラリ
moji = '370-0052 高崎市 旭町，154-8533 東京都世田谷区太子堂，379-2121 群馬県前橋市小屋原町'
yubin = re.findall('[0-9]+-[0-9]+', moji)

```
---

**構文**

re.findall( マッチパターン定義，パターン探索文字列 )

**マッチパターンには文法がある（正規表現)** 
参考サイト：https://techplay.jp/column/463


---

````python
history = "バナナ　１袋　199円 (税込8% 215円) りんご　１コ 199円 (税込8% 215円) アボカド １コ 99円 (税込8% 107円) オレンジ　１コ 139円 (税込8% 150円)"
print(history)
````

````python
import re
tax = re.findall('% [0-9]+円', history)
print(tax)
````

````python
print(tax[1])
````

````python
tax = re.findall('(% )([0-9]+)(円)', history)
print(tax)
````

````python
print(tax[2])
````

````python
print(tax[2][1])
````
